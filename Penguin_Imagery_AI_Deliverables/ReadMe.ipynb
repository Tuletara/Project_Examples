**Penguin Imagery Classification** 

  These notebooks aim to create an automated classification of satellite imagery that identifies the presence of penguin colonies in Antarctica. The training data used includes an image dataset accessed from Kaggle as well as previously identified penguin colony images.

  The penguin data is contained across five hard drives in various naming conventions and therefore must be consolidated into one location and convention. Further processing is needed to clip the imagery to the penguin colony shapefiles, tile them into smaller scenes to feed into the pretrained model, convert tif format to png, label the tiled scenes as “penguin”, “no penguin” by putting them into separate folders and finally splitting them into training and testing data. 

  The classified image training data is downloaded and split into training and testing datasets then loaded into the built model. The model is saved then fine tuned. 

The notebooks are broken into these tasks: 

	Penguin_data_preparation Notebook: This notebook consolidates, clips, labels and tiles satellite imagery and penguin colony shapefiles 

 Penguin_Model Notebook: This will build off of a pretrained model to classify the penguin dataset previously created.
