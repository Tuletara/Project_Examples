```python
# Set basemap to Portugal area
m = folium.Map(location = [37.1624, -8.8166], tiles = 'Stamen Terrain', zoom_start = 7)

# Add study sites
forest_site = gpd.read_file('DATA/forest_study_site.shp')
folium.GeoJson(data = forest_site['geometry']).add_to(m)

nonFor_site = gpd.read_file('DATA/non-forest_site.shp')
folium.GeoJson(data = nonFor_site['geometry']).add_to(m)

# Display map with layers
m
```
