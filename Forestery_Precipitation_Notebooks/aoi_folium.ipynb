```python
import folium
# Set basemap to Portugal area
m = folium.Map(location = [37.1624, -8.8166], tiles = 'Stamen Terrain', zoom_start = 7)

# Add study sites
forest_site = gpd.read_file('/content/sample_data/forest_study_site.shp')
folium.GeoJson(data = forest_site['geometry']).add_to(m)

nonFor_site = gpd.read_file('/content/sample_data/non-forest_site.shp')
folium.GeoJson(data = nonFor_site['geometry']).add_to(m)

# Display map with layers
m
```

```python
# Add the all shapefile files into the sample_data folder each time 

cities_shp = '../data/us-cities.shp'
cities = geemap.shp_to_ee(cities_shp)
Map.addLayer(cities, {}, 'US Cities')
```
